```python
import os
import shutil
import random

def create_10_fold_validation(src_folder, dest_folder):
    # Create the destination folder if it doesn't exist
    if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)

    # Initialize a list to store all image files found in the source folder and its subdirectories
    image_files = []

    # Traverse through the source folder and its subdirectories to find image files
    for root, _, files in os.walk(src_folder):
        image_files.extend([os.path.join(root, f) for f in files if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif'))])

    # Shuffle the list of image files randomly
    random.shuffle(image_files)

    # Calculate the number of images in each fold
    num_images = len(image_files)
    images_per_fold = num_images // 10

    # Create 10-fold validation folders and copy images to them
    for fold_num in range(10):
        start_idx = fold_num * images_per_fold
        end_idx = (fold_num + 1) * images_per_fold

        # The last fold may have a few more images if the total number of images is not divisible evenly
        if fold_num == 9:
            end_idx = num_images

        # Create the destination folder for the current fold
        fold_dest_folder = os.path.join(dest_folder, f"Fold_{fold_num + 1}")
        if not os.path.exists(fold_dest_folder):
            os.makedirs(fold_dest_folder)

        # Copy images to the destination folder, preserving the subfolder structure
        for idx in range(start_idx, end_idx):
            image_file = image_files[idx]
            relative_path = os.path.relpath(image_file, src_folder)
            dest_path = os.path.join(fold_dest_folder, relative_path)
            os.makedirs(os.path.dirname(dest_path), exist_ok=True)
            shutil.copy(image_file, dest_path)

if __name__ == "__main__":
    # Replace 'source_folder' with the path to your folder containing images and its subdirectories
    source_folder = "datasets"
    # Replace 'destination_folder' with the path to the folder where you want to create the 10-fold validation
    destination_folder = "cross_val_test"
    create_10_fold_validation(source_folder, destination_folder)
```

```python
import os
import shutil
import random
from collections import defaultdict

def create_10_fold_validation(src_folder, dest_folder):
    # Create the destination folder if it doesn't exist
    if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)

    # Initialize a dictionary to store image files from each subfolder
    subfolder_images = defaultdict(list)

    # Traverse through the source folder and its subdirectories to find image files
    for root, _, files in os.walk(src_folder):
        for f in files:
            if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif')):
                subfolder_images[root].append(os.path.join(root, f))

    # Shuffle the image lists within each subfolder separately
    for subfolder in subfolder_images.values():
        random.shuffle(subfolder)

    # Calculate the number of images in each fold
    num_images_per_fold = sum(len(images) for images in subfolder_images.values()) // 10

    # Create 10-fold validation folders and copy images to them, maintaining the stratified split
    for fold_num in range(10):
        fold_dest_folder = os.path.join(dest_folder, f"Fold_{fold_num + 1}")
        if not os.path.exists(fold_dest_folder):
            os.makedirs(fold_dest_folder)

        # Initialize the current fold's image list
        fold_images = []

        # Iterate through each subfolder to collect images for the current fold
        for subfolder_images_list in subfolder_images.values():
            while len(fold_images) < num_images_per_fold:
                # Pop an image from the subfolder_images_list and add it to the fold_images list
                if subfolder_images_list:
                    fold_images.append(subfolder_images_list.pop())

        # Copy images to the destination folder for the current fold
        for image_file in fold_images:
            relative_path = os.path.relpath(image_file, src_folder)
            dest_path = os.path.join(fold_dest_folder, relative_path)
            os.makedirs(os.path.dirname(dest_path), exist_ok=True)
            shutil.copy(image_file, dest_path)

if __name__ == "__main__":
    # Replace 'source_folder' with the path to your folder containing images and its subdirectories
    source_folder = "datasets"
    # Replace 'destination_folder' with the path to the folder where you want to create the 10-fold validation
    destination_folder = "cross_val_test"
    create_10_fold_validation(source_folder, destination_folder)
```
