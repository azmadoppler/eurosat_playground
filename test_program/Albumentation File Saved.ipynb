```python
import os
import albumentations as A
from PIL import Image

def augment_images(directory, output_directory):
    # Define augmentation pipeline
    transform = A.Compose([
        A.HorizontalFlip(p=0.5),
        A.VerticalFlip(p=0.5),
        A.RandomRotate90(p=0.5),
        A.RandomBrightnessContrast(p=0.2),
    ])

    # Create the output directory if it doesn't exist
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # Traverse through all directories and subdirectories
    for root, _, files in os.walk(directory):
        for image_file in files:
            # Check if the file is an image (you can modify this condition based on your image file extensions)
            if image_file.endswith(".jpg") or image_file.endswith(".png") or image_file.endswith(".jpeg"):
                # Load the image
                image_path = os.path.join(root, image_file)
                image = Image.open(image_path)

                # Apply augmentation
                augmented_image = transform(image=np.array(image))['image']

                # Save the augmented image
                relative_path = os.path.relpath(root, directory)  # Get the relative path of the subdirectory
                output_subdir = os.path.join(output_directory, relative_path)
                output_path = os.path.join(output_subdir, image_file)
                
                if not os.path.exists(output_subdir):
                    os.makedirs(output_subdir)

                augmented_image = Image.fromarray(augmented_image)
                augmented_image.save(output_path)

    print("Augmentation complete!")


    
# Specify the directory containing the images
#input_directory = "data/pizza_steak_sushi/test/"

# Specify the output directory to save augmented images
#output_directory = "augment/pizza_steak_sushi/"
# Call the function to augment the images
#augment_images(input_directory, output_directory)
```
